```python
import brightway2 as bw
from brightway2 import *
import time
import os
import numpy as np
import pandas as pd
from polyviz import sankey
```

```python
bw.projects
```

```python
startTime = time.time() # just to see how much time the code takes to run (this is the start)
```

```python
bw.projects.set_current('Prospective LCA v2') # set current project
```

```python
databaseNames = bw.databases
myDatabaseNames = []
for databaseName in databaseNames:
    if 'Abhi' in databaseName:
        if 'image' in databaseName:
            if 'Base' in databaseName or 'RCP19' in databaseName:
                if '2020' in databaseName or '2050' in databaseName:
                    myDatabaseNames.append(databaseName)
```

```python
method = ('IPCC 2013', 'climate change', 'GWP 100a, incl. H and bio CO2')
```

```python
for myDatabaseName in myDatabaseNames:

    db = bw.Database(myDatabaseName)

    hydrogenBAU = [activity for activity in db if 'hydrogen, Abhi' in activity['reference product']
                                                and 'hydrogen production without electricity, steam reforming' in activity['name']
                                                and 'GLO' in activity['location']][0]
    fileName = hydrogenBAU['database']
    fileName = fileName.replace('lci-Abhi ', '')
    fileName = fileName.replace('-Base', ' RCP6')
    fileName = fileName.replace('-', ' RCP19')
    sankey(activity = hydrogenBAU,
           level = 5,
           cutoff = 0.05,
           method = method,
           filepath = os.path.join('..', 'Plots', 'Sankey', 'Hydrogen BAU WO Elec ' + fileName + '.html'))
           
    hydrogenSolar = [activity for activity in db if 'hydrogen, Abhi' in activity['reference product']
                                                and 'solar' in activity['name']
                                                and 'GLO' in activity['location']][0]
    fileName = hydrogenSolar['database']
    fileName = fileName.replace('lci-Abhi ', '')
    fileName = fileName.replace('-Base', ' RCP6')
    fileName = fileName.replace('-', ' RCP19')
    sankey(activity = hydrogenSolar,
           level = 5,
           cutoff = 0.02,
           method = method,
           filepath = os.path.join('..', 'Plots', 'Sankey', 'Hydrogen solar ' + fileName + '.html'))

    hydrogenWind = [activity for activity in db if 'hydrogen, Abhi' in activity['reference product']
                                                and 'onshore wind' in activity['name']
                                                and 'GLO' in activity['location']][0]
    fileName = hydrogenWind['database']
    fileName = fileName.replace('lci-Abhi ', '')
    fileName = fileName.replace('-Base', ' RCP6')
    fileName = fileName.replace('-', ' RCP19')
    sankey(activity = hydrogenWind,
           level = 5,
           cutoff = 0.02,
           method = method,
           filepath = os.path.join('..', 'Plots', 'Sankey', 'Hydrogen wind ' + fileName + '.html'))

    ammoniaBAU = [activity for activity in db if 'ammonia, Abhi' in activity['reference product']
                                                and 'ammonia, BAU' in activity['name']
                                                and 'GLO' in activity['location']][0]
    fileName = ammoniaBAU['database']
    fileName = fileName.replace('lci-Abhi ', '')
    fileName = fileName.replace('-Base', ' RCP6')
    fileName = fileName.replace('-', ' RCP19')
    sankey(activity = ammoniaBAU,
        level = 5,
        cutoff = 0.05,
        method = method,
        filepath = os.path.join('..', 'Plots', 'Sankey', 'Ammonia BAU ' + fileName + '.html'))

    ammoniaSolar = [activity for activity in db if 'ammonia, Abhi' in activity['reference product']
                                                and 'solar' in activity['name']
                                                and 'GLO' in activity['location']][0]
    fileName = ammoniaSolar['database']
    fileName = fileName.replace('lci-Abhi ', '')
    fileName = fileName.replace('-Base', ' RCP6')
    fileName = fileName.replace('-', ' RCP19')
    sankey(activity = ammoniaSolar,
        level = 5,
        cutoff = 0.02,
        method = method,
        filepath = os.path.join('..', 'Plots', 'Sankey', 'Ammonia solar ' + fileName + '.html'))

    ammoniaWind = [activity for activity in db if 'Ammonia, Abhi' in activity['reference product']
                                                and 'onshore wind' in activity['name']
                                                and 'GLO' in activity['location']][0]
    fileName = ammoniaWind['database']
    fileName = fileName.replace('lci-Abhi ', '')
    fileName = fileName.replace('-Base', ' RCP6')
    fileName = fileName.replace('-', ' RCP19')
    sankey(activity = ammoniaWind,
        level = 5,
        cutoff = 0.02,
        method = method,
        filepath = os.path.join('..', 'Plots', 'Sankey', 'Ammonia wind ' + fileName + '.html'))

    methanolBAU = [activity for activity in db if 'methanol, Abhi' in activity['reference product']
                                                and 'methanol, BAU' in activity['name']
                                                and 'GLO' in activity['location']][0]
    fileName = methanolBAU['database']
    fileName = fileName.replace('lci-Abhi ', '')
    fileName = fileName.replace('-Base', ' RCP6')
    fileName = fileName.replace('-', ' RCP19')
    sankey(activity = methanolBAU,
        level = 5,
        cutoff = 0.05,
        method = method,
        filepath = os.path.join('..', 'Plots', 'Sankey', 'Methanol BAU ' + fileName + '.html'))

    methanolSolar = [activity for activity in db if 'methanol, Abhi' in activity['reference product']
                                                and 'methanol synthesis, hydrogen production, gaseous, 200 bar, from PEM electrolysis, from solar electricity carbon dioxide, captured from atmosphere, with a solvent-based direct air capture system, 1MtCO2, with heat pump heat, and onshore wind electricity' in activity['name']
                                                and 'GLO' in activity['location']][0]
    fileName = methanolSolar['database']
    fileName = fileName.replace('lci-Abhi ', '')
    fileName = fileName.replace('-Base', ' RCP6')
    fileName = fileName.replace('-', ' RCP19')
    sankey(activity = methanolSolar,
        level = 5,
        cutoff = 0.02,
        method = method,
        filepath = os.path.join('..', 'Plots', 'Sankey', 'Methanol solar ' + fileName + '.html'))

    methanolWind = [activity for activity in db if 'methanol, Abhi' in activity['reference product']
                                                and 'methanol synthesis, hydrogen production, gaseous, 200 bar, from PEM electrolysis, from onshore wind electricity carbon dioxide, captured from atmosphere, with a solvent-based direct air capture system, 1MtCO2, with heat pump heat, and onshore wind electricity' in activity['name']
                                                and 'GLO' in activity['location']][0]
    fileName = methanolWind['database']
    fileName = fileName.replace('lci-Abhi ', '')
    fileName = fileName.replace('-Base', ' RCP6')
    fileName = fileName.replace('-', ' RCP19')
    sankey(activity = methanolWind,
        level = 5,
        cutoff = 0.02,
        method = method,
        filepath = os.path.join('..', 'Plots', 'Sankey', 'Methanol wind ' + fileName + '.html'))
```

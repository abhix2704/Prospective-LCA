```python
import matplotlib.pyplot as plt

# Sample data
labels = ['Category A', 'Category B', 'Category C']
explode = (0.1, 0.1, 0.1)
sizes = [96, 4, 0]
colors = ['#d2555a', '#a5d76e', '#ffc36e']

# Create a pie chart
fig, ax = plt.subplots()
ax.pie(sizes, startangle = 90, explode = explode, colors = colors)

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')  

plt.savefig('current.png', dpi = 1200, bbox_inches = 'tight', format = 'png', transparent = True)
plt.savefig('current.svg', dpi = 1200, bbox_inches = 'tight', format = 'svg', transparent = True)
```
